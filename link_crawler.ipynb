```python
import requests
from bs4 import BeautifulSoup

def proper_link_creater(og_url, url):
    proper_link = url
    if url[0] == "/":
        url_split = og_url.split("/")
        proper_link = "/".join(url_split[0:3]) + url
    return proper_link

def get_all_hrefs_of_pdfs(url, link_printed, page_visited):
    
    page_source = requests.get(url).text
    soup = BeautifulSoup(page_source, 'lxml')
    a_tags = soup.find_all(name="a")
    master_list = []
    page_visited.add(url)
    master_list.append(a_tags)
    added_set = set()
    url_split2 = url.split("/")
    url_split2 = url_split2[2].split(".")[1]
    while len(master_list) > 0:
        for a_tag in master_list[0]:
            try:
                # print(a_tag)
                if not a_tag.has_attr('href'):
                    continue
                if a_tag["href"]=="" or a_tag["href"][0] not in "h/":
                    continue
                
                # print(a_tag)
                proper_link = proper_link_creater(url, a_tag["href"])
                # print(proper_link)
                if proper_link[-3:] == "pdf":
                    
                    if proper_link not in link_printed:
                        link_printed.add(proper_link)
                        print(a_tag.text)
                        print("^^", proper_link)
           
                    
        # for a_tag in master_list[0]:
        #         if not a_tag.has_attr('href'):
        #             continue
        #         if a_tag["href"]=="" or a_tag["href"][0] == "#":
        #             continue
                # print(a_tag)
                # proper_link = proper_link_creater(url, a_tag["href"])
                elif proper_link not in page_visited:
                    page_visited.add(proper_link)
                    page_source_inside = requests.get(proper_link).text
                    soup_inside = BeautifulSoup(page_source_inside, 'lxml')
                    a_tags_new = soup_inside.find_all(name="a")
                    a_tag_list =[]
                    for tag in a_tags_new:
                        url_split = proper_link.split("/")
                        # url_split = "/".join(url_split[0:3])
                        
                        
                        
                        if url_split[2].split(".")[1] != url_split2:
                            continue
                        if not tag.has_attr('href'):
                            continue
                        if tag["href"]=="" or tag["href"][0] not in "h/" :
                            continue
                        if tag["href"] not in added_set:
                            a_tag_list.append(tag)
                            added_set.add(proper_link_creater(proper_link, tag["href"]))
                    master_list.append(a_tag_list)
                    
                    # print(a_tags_new)
            except Exception as e:
                print(e)
        # print(len(master_list))
        master_list.pop(0)
        print("master_list poped")

    return 


start_url = "https://www.bain.com/about/further-global-responsibility/social-impact/impact-report/"
url_split = start_url.split("/")
homelink = "/".join(url_split[0:3])
link_printed = set()
page_visited = set()
# homelink = 
get_all_hrefs_of_pdfs(start_url, link_printed, page_visited)

# soup = BeautifulSoup(requests.get(start_url).text, 'lxml')
# a_tags = soup.find_all(name="a")
# a_tag = a_tags[0]
```
